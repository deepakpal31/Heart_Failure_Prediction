```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Feature Selection
data=pd.read_csv('heart_failure_clinical_records_dataset.csv')

plt.rcParams['figure.figsize']=15,6 
sns.set_style("darkgrid")

#independant feature
x = data.iloc[:, :-1]
#dependant feature
y = data.iloc[:,-1]

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_) 
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(12).plot(kind='barh')
plt.ylabel('Feature Labels')
plt.xlabel('Feature Importances')
plt.title('Comparison of different Feature Importances')
plt.show()
```

time captures the time of the event. That is, the time at which the patient died or were censored.

Using the time column as a feature is wrong. When you deploy your model, no end user will be able to provide you the value of time, since they do not know at what time in the future the patient will die/get censored!
